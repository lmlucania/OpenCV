```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
HAAR_FILE = "/Users/admin/Documents/project/venv/lib/python3.7/site-packages/cv2/data/haarcascade_frontalcatface.xml"
cascade = cv2.CascadeClassifier(HAAR_FILE)
```

```python
import os
file_list = os.listdir("./test_data")
len(file_list)
```

```python
img_list = []
for img in file_list:
    if "cat" in img:
        img_list.append(img)
len(img_list)
```

```python
import random
img_name = random.choice(img_list)
```

```python
file_path = os.path.join("./test_data/", img_name)
img = cv2.imread(file_path)
img_g = cv2.imread(file_path, 0)
```

```python
face = cascade.detectMultiScale(img_g)
```

```python
plt.figure(figsize=(5, 5))
for x, y, w, h in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 255, 255), 4)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
